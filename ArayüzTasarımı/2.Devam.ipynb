```python
import tkinter as tk

form = tk.Tk()
form.title("Entry")
form.geometry("500x300")

entry = tk.Entry()
entry.pack()

entry2 = tk.Entry(show = "*")
entry2.pack()

def verial():
    etiket["text"] = entry2.get()

def Sil():
    entry.delete(0,"end")
    
button = tk.Button(text = "verileri al",fg = "red",bg = "green",command = verial)
button.pack()

buttonSil = tk.Button(text = "verileri sil",fg = "red",bg = "green",command = Sil)
buttonSil.pack()

etiket = tk.Label(text = "Veriler buraya getirilmesi lazım.")
etiket.pack()

form.mainloop()
```

```python
import tkinter as tk
form = tk.Tk()

form.title("checkbutton")
form.geometry("500x500")
x= tk.IntVar()
x.set(0)
def kontrol():
    if x.get() == 0:
        print("Onaylanmadı")
    else:
        print("Onaylandı")

check = tk.Checkbutton(form,text = "onay",fg = "pink",bg = "green",variable = x,command = kontrol)
check.pack()
                       
form.mainloop()                      
                      
```

```python
import tkinter as tk
import time

form = tk.Tk()
form.title("Zaman")
form.geometry("250x250+500+250")
form.config(bg = "light blue")

def zaman():
    zaman_format = time.strftime("%H:%M:%S")
    zmn_label.config(text = zaman_format)
    zmn_label.after(200,zaman)
    
zmn_label = tk.Label(bg = "white",font = "Times 35 bold")
zmn_label.place(x = 35,y = 80)
zaman()

baslik = tk.Label(text = "Dijital Saat Uygulaması" ,font = "Times 15 bold",bg="light blue")
baslik.place(x = 25 ,y = 20 )
form.mainloop()  
```

```python
import tkinter as tk

form = tk.Tk()
form.title("ScrollBar")
form.geometry("250x250+500+250")
form.config(bg = "light blue")

scrol = tk.Scrollbar()
scrol.pack(side = tk.RIGHT,fill=tk.Y)

ListBox = tk.Listbox(form,yscrollcommand = scrol.get())

for i in range(300):
    ListBox.insert("end","merhaba" + str(i))

ListBox.pack(side = tk.LEFT)    
scrol.config(command = ListBox.yview)


form.mainloop()  
```

```python
from tkinter import *

window = Tk()
window.geometry("680x500")

Label(window, text="Top label").pack()

frame = Frame(window)
frame.pack()

listNodes = Listbox(frame, width=20, height=20, font=("Helvetica", 12))
listNodes.pack(side="left", fill="y")

scrollbar = Scrollbar(frame, orient="vertical")
scrollbar.config(command=listNodes.yview)
scrollbar.pack(side="right", fill="y")

listNodes.config(yscrollcommand=scrollbar.set)

for x in range(100):
    listNodes.insert(END, str(x))

Label(window, text="Bottom label").pack()

window.mainloop()
```

```python
import tkinter as tk 
from tkinter import ttk 
from tkinter import scrolledtext 
  
# Creating tkinter main window 
win = tk.Tk() 
win.title("ScrolledText Widget") 

# Creating scrolled text  
# area widget 
text_area = scrolledtext.ScrolledText(win,  
                                      wrap = tk.WORD,  
                                      width = 40,  
                                      height = 10,  
                                      font = ("Times New Roman", 
                                              15)) 
  
text_area.grid(column = 0, pady = 10, padx = 10) 
  
# Placing cursor in the text area 
text_area.focus() 
win.mainloop() 
```

```python
import tkinter as tk
from tkinter.scrolledtext import ScrolledText

root = tk.Tk()

# Make ScrolledText
txtFinalGrade = ScrolledText(root)
txtFinalGrade.pack()

# Add text to ScrolledText
txtFinalGrade.insert('1.0', 'hello world')

# Get text from ScrolledText
from_scrolledtext = txtFinalGrade.get('1.0', 'end')
print('Text inside ScrolledText is: ', from_scrolledtext)

root.mainloop()
```

```python
from tkinter import * #tkinter'ı programımıza dahil ettik
from PIL import ImageTk, Image #resim eklemek için gerekli kütüphaneyi dahil ettik

class Application(object):

    def __init__(self):

        self.main() #uygulama açılacağı zaman çalışacak ilk fonksiyonu belirledik

    def main(self): #ana pencere de ihtiyaç duyacağımız şeyleri kodlayacağız

        self.pencere1 = Tk()

        self.canvas = Canvas(self.pencere1, width=960, height=540) #resmin penceredeki boyutunu belirledik
        self.image = ImageTk.PhotoImage(Image.open("BackGround//1.jpg")) #img klasöründeki bg_1.png adlı resmi değişkene atadık.

        self.canvas.create_image(0,0, anchor=NW, image=self.image) #resmin konumu ile birlikte oluşturmasını sağladık
        self.canvas.pack() #resmi paketleyip pencerede gösterdik.

app = Application() #uygulamanın çalışması için değişkene atayarak başlattık.
mainloop() #pencerenin biz kapatana kadar açık kalmasını sağlamak için sonsuz döngüye aldık.
```
