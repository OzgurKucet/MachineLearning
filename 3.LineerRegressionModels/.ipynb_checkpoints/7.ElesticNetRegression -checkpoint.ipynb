# Elestic net Regresyonu

Amaç hats kareler toplamını minimize eden katayıları
bu katsayılara bir ceza uygulayarak bulmaktır

ElasticNet L1 ve L2 yaklaşımlarını birleştirir
daha etkin bir düzgünleştirme işlemi yapar
birçok şeye karşı daha dirençli.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge,Lasso,ElasticNet
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn import model_selection
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV,LassoCV,ElasticNetCV
```

```python
df = pd.read_csv("Hitters.csv")
df = df.dropna()
dms = pd.get_dummies(df[["League","Division","NewLeague"]])
#Veri setindeki kategorik değişkenler için dummy değişkenler oluşturduk
#One hot encoding yöntemi ile.
y = df["Salary"] #Bağımlı değişkenimiz
X_ = df.drop(["Salary","League","Division","NewLeague"],axis=1).astype("float64")
#Veri setimizdeki bağımlı değişken ve kategorik değişkenlerimizin orjinal hallerini uçurduk
X = pd.concat([X_,dms[["League_N","Division_W","NewLeague_N"]]],axis = 1)
#Bu kategorik değişkenlerin dummy versyonlarını ve diğer bağımsız
#değişken değerlerimizi bir arada tutarak bağımsız değişkenlerimizi
#X olarak oluşturmuş olduk
X_train,X_test,y_train,y_test  =train_test_split(X,y,test_size = 0.25,random_state=42)
```

```python
enet_model = ElasticNet().fit(X_train,y_train)
```

```python
enet_model.coef_
```

```python
enet_model.intercept_
```

```python
#Tahmin
enet_model.predict(X_train)[0:10] #Tahmin edilen değerler
```

```python
enet_model.predict(X_test)[0:10]
```

```python
y_pred = enet_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
r2_score(y_test,y_pred)
```

# Model Tuning

```python
alphas = 10**np.linspace(10,-2,100)*0.5
```

```python
enet_cv_model = ElasticNetCV(alphas = alphas ,cv = 10).fit(X_train,y_train)
```

```python
enet_cv_model.alpha_
```

```python
enet_cv_model.intercept_
```

```python
#ElasticNet napıyor ridge tarzı cezalandırma lasso tarzı değişken seçimi yapıyor
#ikisini birleştiriyor daha iyi bir iş yapar
```

```python
enet_cv_model.coef_
```

```python
#Final modeli
enet_tuned = ElasticNet(alpha = enet_cv_model.alpha_).fit(X_train,y_train)
```

```python
y_pred = enet_tuned.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred)) #test hatamız
```
