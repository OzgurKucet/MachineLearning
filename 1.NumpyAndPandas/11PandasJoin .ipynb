```python
import numpy as np
import pandas as pd

m = np.random.randint(1,30,size = (5,3))
df1 = pd.DataFrame(m,columns = ["var1","var2","var3"])
df1
```

```python
df2 = df1 + 99
```

# İleri Birleştirme işlenleri

```python
import pandas as pd

df1 = pd.DataFrame({'calisanlar':["Ali","Veli","Ayse","Fatma"],
                   'grup':["Muhasebe","Mühendislik","Mühendislik","ik"]})
df1
```

```python
df2 = pd.DataFrame({'calisanlar':["Ali","Veli","Ayse","Fatma"],
                   'İlk_giris':[2010,2009,2014,2019]})
df2
```

```python
pd.merge(df1,df2) #kendisi neye göre birleştireceğini anladı ama bizde söyleyebiliriz
```

```python
pd.merge(df1,df2,on = "calisanlar") #calisanlar üzerinden birleştirildi
```

```python
#Coktan teke
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df4 = pd.DataFrame({"grup":["Muhasebe","Mühendislik","ik"],
                   "mudur":["Caner","Mustafa","Berkecan"]})
df4
```

```python
pd.merge(df3,df4)
```
